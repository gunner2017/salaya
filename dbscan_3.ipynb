```python
from sklearn import preprocessing
```

```python
X_normalized = preprocessing.normalize(stm_oct_df, norm='l2')
```

```python
print(X_normalized)
```

```python
from sklearn.metrics.pairwise import cosine_distances
```

```python
X_cosine_distance = cosine_distances(X_normalized)
```

```python
print(X_cosine_distance)
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.5, min_samples=10,metric='precomputed').fit(X_cosine_distance)
```

```python
print(db.components_)
```

```python
print(db.core_sample_indices_)
```

```python
print(db.labels_)
```
