```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.99, min_samples=5, metric='euclidean', algorithm='auto', leaf_size=30, p=None, n_jobs=1)
```

```python
db.fit(X_normalized)
```

```python
print(db.components_)
```

```python
print(db.core_sample_indices_)
```
